```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('the_berry_farms_sunflower_field.jpeg',cv.IMREAD_REDUCED_COLOR_4)
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.imshow(img)
plt.xticks([]), plt.yticks([])
plt.show()

# create laplacian of gaussian filter
def blob_filter(size, sigma):
    x, y = np.meshgrid(np.arange(-size/2+1, size/2+1), np.arange(-size/2+1, size/2+1))
    blob = (((x**2 + y**2)/ sigma**2) - 2) * (np.exp(-(x**2+y**2) / (2.0*sigma**2))/(2*np.pi*sigma**4)) 
    blob_normalized = (sigma**2) * blob
    return blob_normalized



blob = blob_filter(15, 1.4)

# apply filter to image
filtered_img = cv.filter2D(img, -1, blob)

# plot image
plt.imshow(filtered_img)
plt.xticks([]), plt.yticks([])
plt.show()





```
